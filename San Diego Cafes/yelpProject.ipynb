```python
# Iterate through each column and find neighborhood (suburb) and place it in a list.

neighborhoods = []

def extract_suburb(STREET):
    
    print(STREET)
    
    full_address = STREET + ", San Diego" 

    url = 'https://nominatim.openstreetmap.org/search'

    PARAMS = {
            'q': full_address,
            'format': 'json',
            'addressdetails': 1  
        }

    response = requests.get(url, params=PARAMS)
    response
    page = response.json()
    if page == []:
        return("No page found.")
    else:
         # Extract the suburb
        if 'suburb' in page[0]['address']:
            suburb = page[0]['address']['suburb']
            print(suburb)
            return(suburb)
        else:
            print("Suburb not found.")
            return("Suburb not found.")
    
# Notes
#'5032 Niagra Ave' isn't being found by the API, so I'll have to manually add it in myself
    
neighborhoods = [extract_suburb(i) for i in working_df['street']]
neighborhoods
```
